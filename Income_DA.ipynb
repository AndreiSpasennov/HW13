```python
import numpy as np
import pandas as pd
income = np.array([150, 100, 180, 200, 300, 250, 320, 280, 400, 350, 220, 130])
expens = np.array([155, 10, 18, 20, 30, 25, 32, 28, 405, 370, 70, 54])
income_without_tax = income*0.7
mon = ['Jn', 'Fb', 'Mr', 'Ar', 'My', 'Jun', 'Jul', 'Ag', 'Sp', 'Ot', 'Nv', 'Dc']
data = {"Income_without_tax": income_without_tax, "Expenses": expens}
df = pd.DataFrame(data, index=mon)
df["Saving"] = df["Income_without_tax"] - df["Expenses"]
df["Profit"] = df["Saving"] > 0 
print(df)
import matplotlib.pyplot as plt
plt.plot(mon, income)
plt.xlabel('Income per month')
plt.show
```

```python
plt.bar(mon, df["Saving"], color=['red', 'coral', 'salmon', 'orange', 'yellow', 'green','lightgreen','skyblue', 'blue', 'purple', 'black', 'brown'])
plt.xlabel('Savings per month')
plt.show
```

```python
x = []
y = []
for i in range(11):
    if df["Saving"][i] > 0:
        x.append(df["Saving"][i])
        y.append(mon[i])
plt.pie(x, colors=['red', 'orange', 'yellow', 'green', 'lightgreen', 'skyblue', 'blue', 'purple', 'black', 'brown'])
plt.legend(y)
plt.xlabel('Part of each month saving')
plt.show
```
