```python
import seaborn as sns
crash = sns.load_dataset("car_crashes")
print(crash.iloc[0:6, :])
sns.displot(data=crash, x='total', kde=True)
```

```python
sns.lmplot(data=crash, x='total', y='speeding')
```

```python
sns.lmplot(data=crash, x='total', y='alcohol')
```

```python
sns.relplot(data=crash, x='ins_premium', y='ins_losses', hue='abbrev')
```
